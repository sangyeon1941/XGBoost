```python
import numpy as np
import pandas as pd
import xgboost as xgb
from sklearn.metrics import accuracy_score
from xgboost import XGBClassifier
```

```python
Train_Label_df = []
for i in range(len(Train_Label)):
    if Train_Label[i] == "adware":
        Train_Label_df.append(0)
    elif Train_Label[i] == "downloader":
        Train_Label_df.append(1)
    elif Train_Label[i] == "hacktool":
        Train_Label_df.append(2)
    elif Train_Label[i] == "miner":
        Train_Label_df.append(3)
    elif Train_Label[i] == "normal":
        Train_Label_df.append(4)
    elif Train_Label[i] == "ransomware":
        Train_Label_df.append(5)
    elif Train_Label[i] == "trojan":
        Train_Label_df.append(6)
    elif Train_Label[i] == "virus":
        Train_Label_df.append(7)
    elif Train_Label[i] == "worm":
        Train_Label_df.append(8)
   
```
